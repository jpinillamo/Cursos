# Subplot

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.hist(y)
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(1,2,2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y,'r--')
plt.plot(y,x,'b:')
plt.subplot(2,1,2)
plt.pie(y)
plt.show()
```

```python
plt.subplot(2,3,1)
plt.plot(x,y)
plt.subplot(2,3,2)
plt.triplot(x,y)
plt.subplot(2,3,3)
plt.pie(y)
plt.subplot(2,3,4)
plt.scatter(y,x)
plt.subplot(2,3,5)
plt.boxplot(x)
plt.subplot(2,3,6)
plt.stem(x,y)
plt.show
```

# SubplotS

```python
x = np.linspace(0,10,22)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x,y,'b')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y,'b')
axes[1].plot(y,x,'r')
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1,ncols=2)
ax1.plot(x,y,'b')
ax2.plot(y,x,'r')
```

```python
fig, ax1 = plt.subplots(1,1)
ax1.fill_between()
```

```python
fig, axes = plt.subplots(2,4)
axes[0,0].plot(x,np.sin(x),'b')
axes[0,1].plot(x,np.cos(x),'y')
axes[0,2].plot(x,np.tan(x),'r')
axes[0,3].plot(x,1/np.tan(x),'g')
axes[1,0].plot(x,1/np.cos(x),'y')
axes[1,1].plot(x,1/np.sin(x),'b:')
axes[1,2].plot(x,y,'b:')
axes[1,3].fill_between(x,y)
fig.tight_layout()
```

# Leyendas, titulos, etc.

```python
fig, axes = plt.subplots(1,2,figsize=(10,5))
axes[0].plot(x,y,label='Sin(x)')
axes[0].set_title('Relaci√≥n X con Y')
axes[0].set_xlabel('Eje X')
axes[0].set_ylabel('Eje Y')
axes[0].legend(loc='lower right',bbox_to_anchor=(1.2,1))
axes[1].plot(y,x)
axes[1].set_title('Relaci√≥n Y con X')
axes[1].set_xlabel('Eje Y')
axes[1].set_ylabel('Eje X')
```

```python
import seaborn as sns
```

```python
country = ['India','Japan','Mexico','Colombia','Germany']
population = [1000,800,900,1000,300]
#plt.style.use('Solarize_Light2')
#plt.style.use('bmh')
```

```python
country = ['India', 'Japan', 'Mexico', 'Colombia', 'Germany']
population = [1000, 800, 900, 1000, 300]

country_population = dict(sorted({country[k]:population[k] for k in range(len(country))}.items(), key = lambda i:i[1], reverse=True))
print(country_population)

#Crea un diccionario iterando listas
#{country[k]:population[k] for k in range(len(country))}

#llama a los items del diccionario
#{country[k]:population[k] for k in range(len(country))}.items()

#Utiliza la funci√≥n sorted para ordenarlos - sorted en este caso debe recibir por que lo vas a ordenar y como va  a quedar el orden final
#sorted({country[k]:population[k] for k in range(len(country))}.items(), key = lambda i:i[1])

#Ordenar por values, por la lista de population
#key = lambda i:i[1]

#Que quede primero el mas grande
#reverse=True

#Volverlo diccionario
#dict(todo)
```

```python
plt.bar(country_population.keys(),country_population.values(), width= 0.5,color=sns.color_palette("tab10"))
plt.xticks(np.arange(5),country_population.keys(),rotation=45)
plt.title('Poblaci√≥n por pais')
plt.xlabel('Pais')
plt.ylabel('Poblaci√≥n en 100 millones de personas')
plt.xlim(-1,5)
plt.ylim(0,1050)
plt.show()
```

```python
plt.barh([i for i in country_population.keys()][::-1],[i for i in country_population.values()][::-1])
plt.title('Poblaci√≥n por pais')
plt.xlabel('Pais')
plt.ylabel('Poblaci√≥n en 100 millones de personas')
plt.ylim(-1,5)
plt.xlim(0,1050)
plt.show()
```
