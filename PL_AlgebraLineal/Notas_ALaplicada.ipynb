## Qu es una pseudoinversa de Moore Penrose y cmo calcularla?

```python
A = np.array([[2,3],[5,7],[11,13]])
A
```

```python
U, D, V = np.linalg.svd(A)
print(U)
print('~~~~~')
print(D)
print('~~~~~')
print(V)
```

```python
D_pse = np.zeros((A.shape[0],A.shape[1])).T
D_pse
```

```python
print('Valores a reemplazar en D_pse')
print(D_pse[:D.shape[0],:D.shape[0]])
```

```python
print('Valores que se ponen en D_pse')
print(np.linalg.inv(np.diag(D)))
```

```python
print('D_pse')
D_pse[:D.shape[0],:D.shape[0]] =  np.linalg.inv(np.diag(D))
print(D_pse)
```

```python
A_pse = V.T.dot(D_pse).dot(U.T)
print(A_pse)
```

```python
A_pse_calc = np.linalg.pinv(A)
A_pse_calc
```

```python
print(A_pse.dot(A))
```

```python
plt.plot(xy[:,0],xy[:,1],'.')
plt.show()
```

```python
xy_centrado = xy- np.mean(xy,axis=0)
```

## Preparando el conjunto de imgenes para aplicar PCA

```python
import sklearn.datasets
data= sklearn.datasets.fetch_olivetti_faces()
import pandas as pd
```

```python
data
```

```python
print(data.data.min())
print(data.data.max())
```

```python
fig, ax = plt.subplots(1,2,figsize=(12,12),
                       subplot_kw={'xticks':[],'yticks':[]})
ax[0].imshow(data.data[0].reshape(64, 64),cmap='gray')
ax[1].imshow(data.data[200].reshape(64, 64),cmap='gray')
```

```python
caras = pd.DataFrame([])
for i in data.data[0]:
    print(i)
```

```python
fig, axes = plt.subplots(10, 10, figsize=(15,15),
                        subplot_kw = {'xticks':[],'yticks':[]},
                        gridspec_kw=dict(hspace=0.01, wspace=0.01))
for i,ax in enumerate(axes.flat):
    ax.imshow(data.data[i].reshape(64,64),cmap='gray')
```

```python
from sklearn.decomposition import PCA

```

```python
caras_pca = PCA (n_components=0.80)
caras_pca.fit(data.data)
```

```python
filas = 3
columnas = caras_pca.n_components_ // filas

fig, axes = plt.subplots(filas, columnas, figsize=(12,6),
                        subplot_kw = {'xticks':[],'yticks':[]},
                        gridspec_kw=dict(hspace=0.01, wspace=0.01))
for i,ax in enumerate(axes.flat):
    ax.imshow(caras_pca.components_[i].reshape(64,64),cmap='gray')
```

```python
print(caras_pca.n_components_)
```

```python
componentes = caras_pca.transform(data.data)
proyeccion = caras_pca.inverse_transform(componentes)

fig, axes = plt.subplots(5, 10, figsize=(12,6),
                        subplot_kw = {'xticks':[],'yticks':[]},
                        gridspec_kw=dict(hspace=0.01, wspace=0.01))
for i,ax in enumerate(axes.flat):
    ax.imshow(proyeccion[i].reshape(64,64),cmap='gray')
```
