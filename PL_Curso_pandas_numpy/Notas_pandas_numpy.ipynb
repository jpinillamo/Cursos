## Shape y Reshape
### Shape:
* Indica la forma que tiene un arreglo
* Con qué estrucutra estoy trabajando
* Nos ayuda con esos datos para poder manipularlo mejor
### Reshape:
* Transforma el arreglo mientras se mantengan los elementos

```python
#creamos un array y luego averiguamos cómo esta constituida
shape_arr = np.random.randint(1,11,(3,2))
print(shape_arr)
shape_arr.shape
```

```python
#transformamos el arreglo mientras mantenemos los elementos
shape_arr.reshape(1,6)
```

```python
#otra forma de hacer lo mismo
np.reshape(shape_arr,(1,6))
```

```python
#como lo haría el lenguaje C
np.reshape(shape_arr,(2,3),'C')
```

```python
#como lo haría el lenguaje Fortran
np.reshape(shape_arr,(2,3),'F')
```

```python
#como lo haría de forma aleatroria y como mi sistema lo nota más optimo
np.reshape(shape_arr,(2,3),'A')
```

# Combinar dataframes


## Concat

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']
 })

df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
 'B':['B4','B5','B6','B7'],
 'C':['C4','C5','C6','C7'],
 'D':['D4','D5','D6','D7']
 })
```

```python
pd.concat([df1,df2], ignore_index=True)
```

```python
pd.concat([df1,df2], axis=1)
```

## Merge

```python
izq = pd.DataFrame({ 'key': ['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3'],
 })
der = pd.DataFrame({ 'key': ['k0','k1','k2','k3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3'],
 })
```

```python
izq.merge(der, on='key')
```

```python
izq = pd.DataFrame({ 'key': ['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3'],
 })
der = pd.DataFrame({ 'key_2': ['k0','k1','k2','k3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3'],
 })
```

```python
izq.merge(der, left_on='key',right_on='key_2')
```

```python
izq = pd.DataFrame({ 'key': ['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3'],
 })
der = pd.DataFrame({ 'key_2': ['k0','k1','k2',np.nan],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3'],
 })
```

```python
izq.merge(der, left_on='key',right_on='key_2',how='right')
```

# Join
Index match

```python
izq = pd.DataFrame({ 'A':['A0','A1','A2'],
 'B':['B0','B1','B2'],
 }, index=['k0','k1','k2'])

der = pd.DataFrame({  'C':['C0','C1','C2'],
 'D':['D0','D1','D2']
 }, index=['k0','k2','k3'])
```

```python
izq.join(der,how='inner')
```
